```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw=pd.read_csv('loan_data.csv')
```

```python
raw.drop(['id','home_ownership_cat','income_category','annual_inc','application_type','purpose_cat','interest_payments','grade','dti','application_type_cat','recoveries','loan_condition','term_cat','issue_d','final_d','total_pymnt','total_rec_prncp'],axis=1,inplace=True) 
raw.head()
```

```python
target_values=raw['loan_condition_cat'].values
raw.drop('loan_condition_cat',axis=1,inplace=True)
```

## One Hot Encoding

```python
oh1=pd.get_dummies(raw['home_ownership'])
```

```python
oh2=pd.get_dummies(raw['term'])
```

```python
oh3=pd.get_dummies(raw['purpose'])
```

```python
oh4=pd.get_dummies(raw['region'])
```

```python
final=pd.concat([raw,oh1,oh2,oh3,oh4],axis=1)
final.head()
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
final['year']=l.fit_transform(final['year'])
l=LabelEncoder()
final.head()
```

```python
col_vals=['year', 'emp_length_int', 'income_cat', 'loan_amount',
         'interest_payment_cat', 'interest_rate', 'grade_cat',
       'installment', 'ANY', 'MORTGAGE', 'NONE', 'OTHER', 'OWN',
       'RENT', ' 36 months', ' 60 months', 'car', 'credit_card',
       'debt_consolidation', 'educational', 'home_improvement', 'house',
       'major_purchase', 'medical', 'moving', 'other', 'renewable_energy',
       'small_business', 'vacation', 'wedding', 'Northern-Irl', 'cannught',
       'leinster','munster','ulster']
```

## SMOTE

```python
from imblearn.over_sampling import SMOTE 
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
sm = SMOTE(random_state = 2)
y=target_values
x=final[col_vals]
y=y.reshape(-1, 1)
over = RandomOverSampler(sampling_strategy=0.2)
x, y = over.fit_resample(x,y)
print(x.shape,y.shape)
```

```python
x_array=np.array(x)
df_after_smote= pd.DataFrame(data = x_array,columns = col_vals) 
```

```python
df_after_smote['target']=y
df_after_smote.head()
```
